"getting gold prices"


```
import pandas as pd
pd.core.common.is_list_like=pd.api.types.is_list_like
import pandas_datareader as web
import datetime as dt

stock=['GLD', '^DJI']
start="2000-01-01"
end=dt.datetime.today()
df=web.DataReader(stock,'yahoo',start,end)
```
